# Data augmentation methods

```python
def rotate(image: Image.Image, angle: int):
    pass

def rotate2(image: Image.Image, number_of_rotation: int):
    images = []
    for i in range(number_of_rotation):
        rotation = (360 / number_of_rotation) * i
        new_im = image.rotate(rotation)
        images.append(new_im)
    image = image.transpose(Image.FLIP_LEFT_RIGHT)
    for i in range(number_of_rotation):
        rotation = (360 / number_of_rotation) * i
        new_im = image.rotate(rotation)
        images.append(new_im)
    return images
```

```python
def demultiply2(dataset: pd.DataFrame, images_path: str, nb_rotations: int = 2):
    images = []
    for filename in dataset.filename.values:
        filepath = os.path.join(images_path, filename)
        image = Image.open(filepath)
        new_images = rotate2(image, nb_rotations)
        images.extend(new_images)
    return images
```

```python
def preprocess(dataset: [Image.Image], shape=(224, 224)):
    preprocessed = []
    for image in dataset:
        image = image.resize(shape)
        image = np.array(image)
        image = image[:, :, :3]  # Removing the 4th channel, in case of a .png with transparency
        image = tl_prepro(image)
        preprocessed.append(image)
    return np.asarray(preprocessed)
```

```python
dataset = preprocess(demultiply2(df, KS_IMAGES_PATH))
```
